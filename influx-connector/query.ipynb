```python
import pytz
def convert_to_rome_time(row):
        dt = row['_time']
        rome_tz = pytz.timezone('Europe/Rome')
        rome_time = dt.astimezone(rome_tz)
        row['_time'] = rome_time
        return row

df = df.apply(convert_to_rome_time, axis=1)
```
